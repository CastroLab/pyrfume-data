```python
molecules = pyrfume.load_data('more/odorants/all-cids-properties.csv')
odorant_cids = molecules.index
#odorant_cids = p_odor[p_odor>0.2].index
cid_smiles = molecules.loc[odorant_cids]['SMILES']
```

```python
mordred = smiles_to_mordred(cid_smiles.values)
```

```python
mordred.isnull().sum().sum()
```

```python
nan_counts = mordred.isnull().sum()
good_features = nan_counts[nan_counts<len(nan_counts)*0.25].index
```

```python
ss = StandardScaler()
knn = KNNImputer(n_neighbors=5)
z = ss.fit_transform(mordred[good_features].values)
%time z = knn.fit_transform(z)
z = ss.inverse_transform(z)
mordred_imputed = pd.DataFrame(z, index=mordred.index, columns=mordred[good_features].columns)
```

```python
mordred_imputed.isnull().sum().sum()
```

```python
mordred_imputed = cid_smiles.to_frame(name='SMILES').join(mordred_imputed, on='SMILES', how='inner').drop('SMILES', axis=1)
```
