```python
molecules = pyrfume.load_data('more/odorants/all-cids-properties.csv')
odorant_cids = molecules.index
#odorant_cids = p_odor[p_odor>0.2].index
cid_smiles = molecules.loc[odorant_cids]['SMILES']
```

```python
morgan_sim = smiles_to_morgan_sim(cid_smiles.values, cid_smiles.values)
```

```python
morgan_sim.isnull().sum().sum(), morgan_sim.shape
```

```python
morgan_sim = cid_smiles.to_frame(name='SMILES').join(morgan_sim, on='SMILES', how='inner').drop('SMILES', axis=1)
```
